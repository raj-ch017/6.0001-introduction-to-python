### Problem 1

Assume the string given as input is a string of lower case characters.

Write a program that counts up the number of vowels contained in the string. 
Valid vowels are: 'a', 'e', 'i', 'o', and 'u'. 

**For example**, if string is 'azcbobobegghakl', your program should print:

    Number of vowels: 5

```python
def count_vowels(input_string):
    """
    Parameters:
        input_string (str) - A string sentence
        
    Returns:
        count (int) - Number of vowels in input_string
    """
    
    vowel_list = ['a','e','i','o','u']
    count = 0
    
    for ind in range(len(input_string)):
        
        if input_string[ind] in vowel_list:
            count += 1
            
    return count
```

### Program 2


Write a program that prints the number of times the string 'bob' occurs in a given string. 


**For example**, if s = 'azcbobobegghakl', then your program should print

      Number of times bob occurs is: 2


```python
def sub_string_counter(sub_string,input_string):
    """
    Parameters:
        sub_string (str) - The sub-string to search for
        input_string (str) - The sentence in which to look for sub-string
        
    Returns:
        count (int) - Frequency of sub_string occurences in input_string
    
    """
    
    freq = 0
    
    if sub_string in input_string:
        
        for ind in range(len(input_string)):
            
            if input_string[ind] == 'b':
                
                check = input_string[ind:ind+3]
                
                if check == sub_string:
                    
                    freq += 1
                    
    return freq
```

### Program 3

Write a program that prints the longest substring of a given string. The order to be followed will be the letters occuring in alphabetical order. 

**For example**, if the given string is `'azcbobobegghakl'`, then your program should print:

    Longest substring in alphabetical order is: beggh

In the case of ties, print the first substring. 

**For example**, if string = `'abcbcd'`, then your program should print

    Longest substring in alphabetical order is: abc

```python
def longest_substring(input_string):
    """
    Parameters:
        input_string (str) - The string sentence
        
    Returns:
        out_string (str) - The longest substring in input_string
    """
    
    out_string = ''
    substring_len = []
    count = 1

    for x in range(len(input_string)-1):
        
        if ord(input_string[x+1]) >= ord(input_string[x]):
            count += 1 
        
        else:
            substring_len.append(count)
            count = 1



    if count == len(input_string):
        return input_string
    
    elif count > max(substring_len):
        
        substring_len.append(count)
        max_val = max(substring_len)
        max_index = substring_len.index(max_val)
        total = 0
        
        for z in range(max_index):
            total = total + substring_len[z]
        
        out_string = input_string[total:total+max_val]
        return out_string
    
    else:
        
        max_val = max(substring_len)
        max_index = substring_len.index(max_val)
        total = 0
        
        for z in range(max_index):
            total = total + substring_len[z]
        
        out_string = input_string[total:total+max_val]
        return out_string
```
